# Paquete NumPy

NumPy es el paquete bÃ¡sico para computaciÃ³n cientÃ­fica en Python. Proporciona arrays multidimensionales, colecciones de funciones optimizadas para arrays: matemÃ¡ticas, lÃ³gicas, manipulaciÃ³n forma, ordenaciÃ³n, selecciÃ³n, I/O, FFT, Ã¡lgebra lineal bÃ¡sica, operaciones bÃ¡sicas de estadÃ­stica, ramdom y mucho mÃ¡s.

El core de NumPy es el objeto *ndarray*, encapsula un array de n-dimensiones de datos de tÃ­pos homogÃ©neos con muchas operaciones compiladas para un buen rendimiento.

CaracterÃ­sticas arrays NumPy:

* Tienen un tamaÃ±o fijo al crearse. Para cambiar de tamaÃ±o se crea un nuevo array y se elimina el original.
* Los elementos del array tienen que tener el mismo tipo, por tanto el mismo tamaÃ±o en memoria. Hay una excepciÃ³n con los arrays de objetos que permiten tener arrays con elementos de diferentes tamaÃ±os.
* Facilita las operaciones matemÃ¡ticas de un grÃ¡n nÃºmero de datos. Por lo general, este tipo de operaciones se ejecutan de manera mÃ¡s eficiente y con menos cÃ³digo.
* Una comunidad creciente de cientÃ­ficos y matemÃ¡ticos estÃ¡n basando sus paquetes python en NumPy arrays.
* Permite **VectorizaciÃ³n**:
 * Ausencia de buches explÃ­citos
 * Permite optimizaciÃ³n cercana a cÃ³digo C
 * Es mÃ¡s fÃ¡cil de leer
 * Menos lÃ­neas de cÃ³digo generalmente significa menos bugs
 * El cÃ³digo es mÃ¡s parecido a la notaciÃ³n matemÃ¡tica
* Permite **Broadcasting**:
 * AplicaciÃ³n de operaciones implÃ­citas elemento a elemento

# CreaciÃ³n de matrices

```python
matriz = np.zeros((2, 3))
print matriz
print "Tipo elemento por defecto:", type (matriz[0,0]), ", Size elemento:", matriz[0,0].nbytes
print "NÃºmero de dimensiones: ", matriz.ndim, "TamaÃ±o matriz: ", matriz.shape, "NÃºmero elementos: ", matriz.size
print "Tipo de elementos: ", matriz.dtype, "TamaÃ±o elementos: ", matriz.itemsize, "Bytes"
```

```python
matriz = np.ones((2, 3), dtype=np.int32)
print matriz
print "Tipo elemento por defecto:", type (matriz[0,0]), ", Size elemento:", matriz[0,0].nbytes
```
